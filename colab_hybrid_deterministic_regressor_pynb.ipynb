```python
import os
import zipfile
import warnings
import cv2
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, backend as K
from sklearn.model_selection import LeaveOneGroupOut
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from google.colab import files
from tqdm.notebook import tqdm

# --- Comprehensive Reproducibility Block ---
# This ensures the results are as stable as possible for a given environment.
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
np.random.seed(SEED)
tf.random.set_seed(SEED)
tf.config.experimental.enable_op_determinism()

# --- Setup ---
warnings.filterwarnings('ignore', category=UserWarning)
sns.set_theme(style="whitegrid", palette="colorblind")
print("Similarity Learning with Siamese Network for Material Identification")
print("-" * 60)

# --- Configuration ---
ZIP_FILE_PATH = "/content/drive/MyDrive/roughness images.zip"
CSV_FILE_PATH = "dataset_roughness.csv"
EXTRACTION_DIR = "extracted_data"
IMG_SIZE = 128
MARGIN = 1.0
EPOCHS = 2 # A sufficient number of epochs for the model to learn
BATCH_SIZE = 32
LEARNING_RATE = 0.0001 # The lower learning rate that helps prevent getting stuck

# ==============================================================================
# Step 1: File Upload and Extraction
# ==============================================================================
print("\n--- Step 1: Uploading and Extracting Data ---")

if not os.path.exists(ZIP_FILE_PATH):
    print(f"'{ZIP_FILE_PATH}' not found. Please check the path in your Google Drive.")

if not os.path.exists(CSV_FILE_PATH):
    print(f"\nPlease upload '{CSV_FILE_PATH}' to the Colab session.")
    uploaded_csv = files.upload()
    if not uploaded_csv: exit("Upload cancelled. Halting execution.")

print(f"\nExtracting '{ZIP_FILE_PATH}'...")
with zipfile.ZipFile(ZIP_FILE_PATH, 'r') as zip_ref:
    zip_ref.extractall(EXTRACTION_DIR)
print("--> Extraction complete.")

DATA_DIR = os.path.join(EXTRACTION_DIR, "roughness images")
if not os.path.isdir(DATA_DIR):
    extracted_contents = [d for d in os.listdir(EXTRACTION_DIR) if os.path.isdir(os.path.join(EXTRACTION_DIR, d))]
    if extracted_contents:
        DATA_DIR = os.path.join(EXTRACTION_DIR, extracted_contents[0])

# ==============================================================================
# Step 2: Data Loading and Preparation
# ==============================================================================
print("\n--- Step 2: Loading and Preparing Data ---")

df_roughness = pd.read_csv(CSV_FILE_PATH)
df_roughness.rename(columns={
    'Sample no.': 'Sample', 'Line': 'Image_Index',
    'Deposition Thickness': 'Deposition'
}, inplace=True, errors='ignore')

all_images_info = []
for mat in ["pla", "tpu"]:
    mat_dir = os.path.join(DATA_DIR, mat)
    if not os.path.isdir(mat_dir): continue
    for sample_dir in sorted(os.listdir(mat_dir)):
        try: sample_id = int(sample_dir.split('.')[0])
        except ValueError: continue
        full_sample_dir = os.path.join(mat_dir, sample_dir)
        if not os.path.isdir(full_sample_dir): continue
        image_files = sorted([f for f in os.listdir(full_sample_dir) if f.lower().endswith((".png", ".jpg", ".jpeg"))])
        for img_file in image_files:
            all_images_info.append({
                "Material": mat.upper(), "Sample": sample_id,
                "image_path": os.path.join(full_sample_dir, img_file)
            })

df_final = pd.DataFrame(all_images_info)
df_final['unique_sample_id'] = df_final['Material'].astype(str) + "_" + df_final['Sample'].astype(str)

print(f"--> Found {len(df_final)} images from {len(df_final['unique_sample_id'].unique())} unique samples.")

# ==============================================================================
# Step 3: Siamese Network and LOOCV
# ==============================================================================
print("\n--- Step 3: Performing LOOCV with Siamese Network ---")

def create_image_pairs(df, total_pairs):
    """Creates a balanced set of positive (within-sample) and negative pairs."""
    pairs, labels = [], []

    # --- MODIFIED: Reverted to "within-sample" positive pair creation ---
    # Group images by their unique sample ID
    grouped = df.groupby('unique_sample_id')

    print(" -> Creating positive (within-sample) pairs...")
    # Create positive pairs by taking all combinations within each sample group
    for _, group_df in tqdm(grouped, desc="Processing Samples for Positives"):
        # Create all unique pairs of image paths within this sample
        for i, j in itertools.combinations(group_df['image_path'], 2):
            pairs.append([i, j])
            labels.append(1) # Label 1 for "same"

    # --- Negative pair creation remains the same, but we balance it ---
    pla_paths = df[df['Material'] == 'PLA']['image_path'].tolist()
    tpu_paths = df[df['Material'] == 'TPU']['image_path'].tolist()

    # Create a balanced number of negative pairs (equal to the number of positive pairs)
    num_neg_pairs = len(labels)

    print(" -> Creating negative (across-material) pairs...")
    # Create negative pairs (different materials)
    for _ in tqdm(range(num_neg_pairs), desc="Creating Negative Pairs"):
        if not pla_paths or not tpu_paths: continue
        path1 = np.random.choice(pla_paths)
        path2 = np.random.choice(tpu_paths)
        pairs.append([path1, path2]); labels.append(0)

    return np.array(pairs), np.array(labels)

def load_and_preprocess_image(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img.astype('float32') / 255.0
    return np.expand_dims(img, axis=-1)

def euclidean_distance(vectors):
    (featsA, featsB) = vectors
    sumSquared = K.sum(K.square(featsA - featsB), axis=1, keepdims=True)
    return K.sqrt(K.maximum(sumSquared, K.epsilon()))

def contrastive_loss(y, preds):
    squaredPreds = K.square(preds)
    squaredMargin = K.square(K.maximum(MARGIN - preds, 0))
    loss = K.mean(y * squaredPreds + (1 - y) * squaredMargin)
    return loss

def build_siamese_encoder(input_shape):
    return models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32, (3, 3), activation='relu'), layers.BatchNormalization(), layers.MaxPooling2D(),
        layers.Conv2D(64, (3, 3), activation='relu'), layers.BatchNormalization(), layers.MaxPooling2D(),
        layers.Flatten(), layers.Dense(128, activation='relu')
    ])

# --- Main LOOCV Loop ---
groups = df_final['unique_sample_id'].values
unique_samples = sorted(df_final['unique_sample_id'].unique())
logo = LeaveOneGroupOut()
all_true_labels, all_pred_labels = [], []

for fold, (train_indices, val_indices) in tqdm(enumerate(logo.split(df_final, groups=groups)), total=len(unique_samples), desc="Overall Progress"):
    val_sample_id = unique_samples[fold]
    print(f"\n--- FOLD {fold+1}/{len(unique_samples)}: Validating on Sample {val_sample_id} ---")

    df_train, df_val = df_final.iloc[train_indices], df_final.iloc[val_indices]

    num_training_images = len(df_train)
    total_pairs_to_create = num_training_images * 10

    print(f" -> Creating ~{total_pairs_to_create} training pairs...")
    train_pairs, train_labels = create_image_pairs(df_train, total_pairs_to_create)

    encoder = build_siamese_encoder((IMG_SIZE, IMG_SIZE, 1))
    imgA = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 1)); imgB = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 1))
    featA = encoder(imgA); featB = encoder(imgB)
    distance = layers.Lambda(euclidean_distance)([featA, featB])
    siamese_model = models.Model(inputs=[imgA, imgB], outputs=distance)

    # Use the Adam optimizer with a lower learning rate for stable training
    optimizer = tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE)
    siamese_model.compile(loss=contrastive_loss, optimizer=optimizer)

    print(" -> Preparing training data...")
    pair_images_A = np.array([load_and_preprocess_image(p[0]) for p in tqdm(train_pairs, desc="Loading Pair A")])
    pair_images_B = np.array([load_and_preprocess_image(p[1]) for p in tqdm(train_pairs, desc="Loading Pair B")])

    print(" -> Training Siamese model...")
    siamese_model.fit([pair_images_A, pair_images_B], train_labels,
                      epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=1, shuffle=True)

    print(" -> Creating anchor embeddings...")
    pla_anchor_paths = df_train[df_train['Material'] == 'PLA']['image_path']
    tpu_anchor_paths = df_train[df_train['Material'] == 'TPU']['image_path']

    pla_embeddings = encoder.predict(np.array([load_and_preprocess_image(p) for p in tqdm(pla_anchor_paths, desc="PLA Anchors")]), verbose=0)
    tpu_embeddings = encoder.predict(np.array([load_and_preprocess_image(p) for p in tqdm(tpu_anchor_paths, desc="TPU Anchors")]), verbose=0)

    anchor_pla = np.mean(pla_embeddings, axis=0)
    anchor_tpu = np.mean(tpu_embeddings, axis=0)

    val_image_paths = df_val['image_path']
    val_embeddings = encoder.predict(np.array([load_and_preprocess_image(p) for p in val_image_paths]), verbose=0)
    avg_val_embedding = np.mean(val_embeddings, axis=0)

    dist_to_pla = np.linalg.norm(avg_val_embedding - anchor_pla)
    dist_to_tpu = np.linalg.norm(avg_val_embedding - anchor_tpu)

    predicted_material = 'PLA' if dist_to_pla < dist_to_tpu else 'TPU'
    true_material = df_val['Material'].iloc[0]

    all_true_labels.append(true_material)
    all_pred_labels.append(predicted_material)
    print(f" -> True: {true_material}, Predicted: {predicted_material} (Dist PLA: {dist_to_pla:.4f}, Dist TPU: {dist_to_tpu:.4f})")

# ==============================================================================
# Step 4: Final Evaluation
# ==============================================================================
print("\n\n--- Step 4: Final LOOCV Evaluation ---")

df_results = pd.DataFrame({'sample_id': unique_samples, 'true_material': all_true_labels, 'pred_material': all_pred_labels})
print("\nFinal Predictions from LOOCV:")
print(df_results)

print("\n--- Overall Classification Report ---")
report = classification_report(df_results['true_material'], df_results['pred_material'], zero_division=0)
print(report)

print("\n--- Overall Confusion Matrix ---")
le = LabelEncoder().fit(df_final['Material'].unique())
cm = confusion_matrix(df_results['true_material'], df_results['pred_material'], labels=le.classes_)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Overall Confusion Matrix (Siamese Network LOOCV)')
plt.xlabel('Predicted Label'); plt.ylabel('True Label')
plt.show()

print("\nScript finished successfully.")

```

```python
import os
import warnings
import numpy as np
import pandas as pd
import cv2
import zipfile
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, backend as K
from sklearn.model_selection import GroupKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.cluster import KMeans
from google.colab import files
from tqdm.notebook import tqdm

# --- Setup ---
warnings.filterwarnings('ignore', category=UserWarning)
sns.set_theme(style="whitegrid", palette="colorblind")
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("Hybrid Unsupervised Clustering + Supervised Regressor")
print("-" * 60)

# --- Configuration ---
CSV_FILE_PATH = "dataset_roughness.csv"
ZIP_FILE_PATH = "/content/drive/MyDrive/roughness images.zip"
EXTRACTION_DIR = "extracted_data"
IMG_SIZE = 128
ENCODER_EPOCHS = 2
REGRESSOR_EPOCHS = 200
BATCH_SIZE = 32
LEARNING_RATE = 0.0001
N_SPLITS = 5 # Use 5 splits for GroupKFold

# ==============================================================================
# Step 1: File Upload and Data Loading
# ==============================================================================
print("\n--- Step 1: Uploading and Loading Data ---")
if not os.path.exists(CSV_FILE_PATH):
    print(f"Please upload '{CSV_FILE_PATH}'.")
    uploaded_csv = files.upload()
    if not uploaded_csv: exit("Upload cancelled.")
if not os.path.exists(ZIP_FILE_PATH):
    print(f"'{ZIP_FILE_PATH}' not found. Please check path.")
else:
    print(f"\nExtracting '{ZIP_FILE_PATH}'...")
    with zipfile.ZipFile(ZIP_FILE_PATH, 'r') as zip_ref: zip_ref.extractall(EXTRACTION_DIR)
    print("--> Extraction complete.")
DATA_DIR = os.path.join(EXTRACTION_DIR, "roughness images")
if not os.path.isdir(DATA_DIR):
    extracted_contents = [d for d in os.listdir(EXTRACTION_DIR) if os.path.isdir(os.path.join(EXTRACTION_DIR, d))]
    if extracted_contents: DATA_DIR = os.path.join(EXTRACTION_DIR, extracted_contents[0])

df_roughness = pd.read_csv(CSV_FILE_PATH)
df_roughness.rename(columns={'Sample no.': 'Sample', 'Line': 'Image_Index', 'Deposition Thickness': 'Deposition'}, inplace=True, errors='ignore')

dataset_info = []
for mat in ["pla", "tpu"]:
    mat_dir = os.path.join(DATA_DIR, mat)
    if not os.path.isdir(mat_dir): continue
    for sample_dir in sorted(os.listdir(mat_dir)):
        try: sample_id = int(sample_dir.split('.')[0])
        except ValueError: continue
        full_sample_dir = os.path.join(mat_dir, sample_dir)
        if not os.path.isdir(full_sample_dir): continue
        image_files = sorted([f for f in os.listdir(full_sample_dir) if f.lower().endswith((".png", ".jpg", ".jpeg"))])
        if len(image_files) >= 10:
            selected_files = image_files[:10]
            for i, img_file in enumerate(selected_files):
                dataset_info.append({"Material": mat.upper(), "Sample": sample_id, "Image_Index": i + 1, "image_path": os.path.join(full_sample_dir, img_file)})

df_found_images = pd.DataFrame(dataset_info)
df_final = pd.merge(df_found_images, df_roughness, on=["Material", "Sample", "Image_Index"], how="inner")
df_final['unique_sample_id'] = df_final['Material'].astype(str) + "_" + df_final['Sample'].astype(str)
print(f"--> Loaded and matched {len(df_final)} images from {len(df_final['unique_sample_id'].unique())} unique samples.")

# ==============================================================================
# Stage 1: Unsupervised Visual Clustering
# ==============================================================================
print("\n--- Stage 1: Training Siamese Encoder and Performing Unsupervised Clustering ---")

def create_image_pairs(df, total_pairs):
    pairs, labels = [], []
    pla_paths = df[df['Material'] == 'PLA']['image_path'].tolist()
    tpu_paths = df[df['Material'] == 'TPU']['image_path'].tolist()
    for _ in range(total_pairs // 2):
        if np.random.rand() > 0.5 and len(pla_paths) > 1: path1, path2 = np.random.choice(pla_paths, 2, replace=False)
        elif len(tpu_paths) > 1: path1, path2 = np.random.choice(tpu_paths, 2, replace=False)
        else: continue
        pairs.append([path1, path2]); labels.append(1)
    for _ in range(total_pairs // 2):
        if not pla_paths or not tpu_paths: continue
        path1 = np.random.choice(pla_paths); path2 = np.random.choice(tpu_paths)
        pairs.append([path1, path2]); labels.append(0)
    return np.array(pairs), np.array(labels)

def load_and_preprocess_image(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    return np.expand_dims(img.astype('float32') / 255.0, -1)

def contrastive_loss(y, preds, margin=1.0):
    return K.mean(y * K.square(preds) + (1 - y) * K.square(K.maximum(margin - preds, 0)))

def build_siamese_encoder(input_shape):
    return models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32, (3, 3), activation='relu'), layers.BatchNormalization(), layers.MaxPooling2D(),
        layers.Conv2D(64, (3, 3), activation='relu'), layers.BatchNormalization(), layers.MaxPooling2D(),
        layers.Flatten(), layers.Dense(128, activation='relu')], name="encoder")

# Train one encoder on ALL data to learn the best possible feature space
total_pairs = len(df_final) * 20
train_pairs, train_labels = create_image_pairs(df_final, total_pairs)
encoder = build_siamese_encoder((IMG_SIZE, IMG_SIZE, 1))
imgA = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 1)); imgB = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 1))
featA = encoder(imgA); featB = encoder(imgB)
distance = layers.Lambda(lambda x: K.sqrt(K.maximum(K.sum(K.square(x[0] - x[1]), axis=1, keepdims=True), K.epsilon())))([featA, featB])
siamese_model = models.Model(inputs=[imgA, imgB], outputs=distance)
siamese_model.compile(loss=contrastive_loss, optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE))
pair_images_A = np.array([load_and_preprocess_image(p[0]) for p in train_pairs])
pair_images_B = np.array([load_and_preprocess_image(p[1]) for p in train_pairs])
siamese_model.fit([pair_images_A, pair_images_B], train_labels, epochs=ENCODER_EPOCHS, batch_size=BATCH_SIZE, verbose=1)

# Generate embeddings and run K-Means
all_image_data = np.array([load_and_preprocess_image(p) for p in df_final['image_path']])
all_embeddings = encoder.predict(all_image_data, batch_size=BATCH_SIZE, verbose=0)
kmeans = KMeans(n_clusters=2, random_state=SEED, n_init=10)
df_final['visual_cluster'] = kmeans.fit_predict(all_embeddings)
print("--> Unsupervised clustering complete. 'visual_cluster' ID added.")

# ==============================================================================
# Stage 2: Supervised Regression with Fused Features
# ==============================================================================
print("\n--- Stage 2: Training Informed Regressor with GroupKFold CV ---")

# Prepare the final dataset for regression
feature_cols = ['Temperature', 'Printing Speed', 'Deposition', 'visual_cluster']
target_cols = ['Ra'] # Focusing on Ra as the primary target
X = df_final[feature_cols].values
y = df_final[target_cols].values
groups = df_final['unique_sample_id'].values

all_true_values, all_pred_values = [], []
gkf = GroupKFold(n_splits=N_SPLITS)

for fold, (train_index, test_index) in enumerate(gkf.split(X, y, groups=groups)):
    print(f"\n--- FOLD {fold + 1}/{N_SPLITS} ---")
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    scaler_X = StandardScaler(); scaler_y = StandardScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)
    y_train_scaled = scaler_y.fit_transform(y_train)

    model = models.Sequential([
        layers.Input(shape=(len(feature_cols),)),
        layers.Dense(32, activation='relu'),
        layers.Dense(64, activation='relu'),
        layers.Dense(32, activation='relu'),
        layers.Dense(len(target_cols))])
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
    model.compile(loss='mse', optimizer=optimizer)

    print(" -> Training regressor...")
    model.fit(X_train_scaled, y_train_scaled, epochs=REGRESSOR_EPOCHS, batch_size=BATCH_SIZE,
              callbacks=[tf.keras.callbacks.EarlyStopping(patience=15, restore_best_weights=True)],
              verbose=0)

    y_pred_scaled = model.predict(X_test_scaled, verbose=0)
    y_pred_orig = scaler_y.inverse_transform(y_pred_scaled)

    all_true_values.extend(y_test); all_pred_values.extend(y_pred_orig)
    print(" -> Fold complete.")

# ==============================================================================
# Final Evaluation
# ==============================================================================
print("\n\n--- Final Cross-Validation Evaluation ---")
y_true_final = np.array(all_true_values)
y_pred_final = np.array(all_pred_values)

r2 = r2_score(y_true_final, y_pred_final)
mae = mean_absolute_error(y_true_final, y_pred_final)

print("\n--- Overall Regression Performance ---")
print(f"Ra Prediction -> R-squared (R): {r2:.4f}")
print(f"Ra Prediction -> Mean Absolute Error (MAE): {mae:.4f}")

plt.figure(figsize=(8, 8))
plt.scatter(y_true_final, y_pred_final, alpha=0.7)
plt.plot([y_true_final.min(), y_true_final.max()], [y_true_final.min(), y_true_final.max()], 'r--', lw=2)
plt.title(f'Ra Prediction (Hybrid Model) | R = {r2:.3f}')
plt.xlabel("True Ra"); plt.ylabel("Predicted Ra")
plt.grid(True); plt.axis('equal')
plt.show()

print("\nScript finished successfully.")
```

```python
import os
import warnings
import numpy as np
import pandas as pd
import cv2
import zipfile
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, backend as K
from sklearn.model_selection import GroupKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
from google.colab import files
from tqdm.notebook import tqdm

# --- Setup ---
warnings.filterwarnings('ignore', category=UserWarning)
sns.set_theme(style="whitegrid", palette="colorblind")
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("Chained Regression Model (Hybrid NN for Ra -> Linear for Rz)")
print("-" * 60)

# --- Configuration ---
CSV_FILE_PATH = "dataset_roughness.csv"
ZIP_FILE_PATH = "/content/drive/MyDrive/roughness images.zip"
EXTRACTION_DIR = "extracted_data"
IMG_SIZE = 128
ENCODER_EPOCHS = 2
REGRESSOR_EPOCHS = 200
BATCH_SIZE = 32
LEARNING_RATE = 0.0001
N_SPLITS = 5

# ==============================================================================
# Step 1: File Upload and Data Loading
# ==============================================================================
print("\n--- Step 1: Uploading and Loading Data ---")
if not os.path.exists(CSV_FILE_PATH):
    print(f"Please upload '{CSV_FILE_PATH}'.")
    uploaded_csv = files.upload()
    if not uploaded_csv: exit("Upload cancelled.")
if not os.path.exists(ZIP_FILE_PATH):
    print(f"'{ZIP_FILE_PATH}' not found. Please check path.")
else:
    print(f"\nExtracting '{ZIP_FILE_PATH}'...")
    with zipfile.ZipFile(ZIP_FILE_PATH, 'r') as zip_ref: zip_ref.extractall(EXTRACTION_DIR)
    print("--> Extraction complete.")
DATA_DIR = os.path.join(EXTRACTION_DIR, "roughness images")
if not os.path.isdir(DATA_DIR):
    extracted_contents = [d for d in os.listdir(EXTRACTION_DIR) if os.path.isdir(os.path.join(EXTRACTION_DIR, d))]
    if extracted_contents: DATA_DIR = os.path.join(EXTRACTION_DIR, extracted_contents[0])

df_roughness = pd.read_csv(CSV_FILE_PATH)
df_roughness.rename(columns={'Sample no.': 'Sample', 'Line': 'Image_Index', 'Deposition Thickness': 'Deposition'}, inplace=True, errors='ignore')

dataset_info = []
for mat in ["pla", "tpu"]:
    mat_dir = os.path.join(DATA_DIR, mat)
    if not os.path.isdir(mat_dir): continue
    for sample_dir in sorted(os.listdir(mat_dir)):
        try: sample_id = int(sample_dir.split('.')[0])
        except ValueError: continue
        full_sample_dir = os.path.join(mat_dir, sample_dir)
        if not os.path.isdir(full_sample_dir): continue
        image_files = sorted([f for f in os.listdir(full_sample_dir) if f.lower().endswith((".png", ".jpg", ".jpeg"))])
        if len(image_files) >= 10:
            selected_files = image_files[:10]
            for i, img_file in enumerate(selected_files):
                dataset_info.append({"Material": mat.upper(), "Sample": sample_id, "Image_Index": i + 1, "image_path": os.path.join(full_sample_dir, img_file)})

df_found_images = pd.DataFrame(dataset_info)
df_final = pd.merge(df_found_images, df_roughness, on=["Material", "Sample", "Image_Index"], how="inner")
df_final['unique_sample_id'] = df_final['Material'].astype(str) + "_" + df_final['Sample'].astype(str)
print(f"--> Loaded and matched {len(df_final)} images from {len(df_final['unique_sample_id'].unique())} unique samples.")

# ==============================================================================
# Stage 1: Unsupervised Visual Clustering
# ==============================================================================
print("\n--- Stage 1: Training Siamese Encoder and Performing Unsupervised Clustering ---")

def create_image_pairs(df, total_pairs):
    pairs, labels = [], []
    pla_paths = df[df['Material'] == 'PLA']['image_path'].tolist()
    tpu_paths = df[df['Material'] == 'TPU']['image_path'].tolist()
    for _ in range(total_pairs // 2):
        if np.random.rand() > 0.5 and len(pla_paths) > 1: path1, path2 = np.random.choice(pla_paths, 2, replace=False)
        elif len(tpu_paths) > 1: path1, path2 = np.random.choice(tpu_paths, 2, replace=False)
        else: continue
        pairs.append([path1, path2]); labels.append(1)
    for _ in range(total_pairs // 2):
        if not pla_paths or not tpu_paths: continue
        path1 = np.random.choice(pla_paths); path2 = np.random.choice(tpu_paths)
        pairs.append([path1, path2]); labels.append(0)
    return np.array(pairs), np.array(labels)

def load_and_preprocess_image(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    return np.expand_dims(img.astype('float32') / 255.0, -1)

def contrastive_loss(y, preds, margin=1.0):
    return K.mean(y * K.square(preds) + (1 - y) * K.square(K.maximum(margin - preds, 0)))

def build_siamese_encoder(input_shape):
    return models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32, (3, 3), activation='relu'), layers.BatchNormalization(), layers.MaxPooling2D(),
        layers.Conv2D(64, (3, 3), activation='relu'), layers.BatchNormalization(), layers.MaxPooling2D(),
        layers.Flatten(), layers.Dense(128, activation='relu')], name="encoder")

total_pairs = len(df_final) * 20
train_pairs, train_labels = create_image_pairs(df_final, total_pairs)
encoder = build_siamese_encoder((IMG_SIZE, IMG_SIZE, 1))
imgA = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 1)); imgB = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 1))
featA = encoder(imgA); featB = encoder(imgB)
distance = layers.Lambda(lambda x: K.sqrt(K.maximum(K.sum(K.square(x[0] - x[1]), axis=1, keepdims=True), K.epsilon())))([featA, featB])
siamese_model = models.Model(inputs=[imgA, imgB], outputs=distance)
siamese_model.compile(loss=contrastive_loss, optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE))
pair_images_A = np.array([load_and_preprocess_image(p[0]) for p in train_pairs])
pair_images_B = np.array([load_and_preprocess_image(p[1]) for p in train_pairs])
siamese_model.fit([pair_images_A, pair_images_B], train_labels, epochs=ENCODER_EPOCHS, batch_size=BATCH_SIZE, verbose=1)

all_image_data = np.array([load_and_preprocess_image(p) for p in df_final['image_path']])
all_embeddings = encoder.predict(all_image_data, batch_size=BATCH_SIZE, verbose=0)
kmeans = KMeans(n_clusters=2, random_state=SEED, n_init=10)
df_final['visual_cluster'] = kmeans.fit_predict(all_embeddings)
print("--> Unsupervised clustering complete. 'visual_cluster' ID added.")

# ==============================================================================
# Stage 2: Supervised Chained Regression
# ==============================================================================
print("\n--- Stage 2: Training Chained Regressor with GroupKFold CV ---")

feature_cols = ['Temperature', 'Printing Speed', 'Deposition', 'visual_cluster']
target_cols_ra = ['Ra'] # The specialist model only predicts Ra
X = df_final[feature_cols].values
y_ra = df_final[target_cols_ra].values
y_rz = df_final[['Rz']].values # We need this for the final evaluation
groups = df_final['unique_sample_id'].values

all_true_ra, all_pred_ra = [], []
all_true_rz = [] # Store the corresponding true Rz values
gkf = GroupKFold(n_splits=N_SPLITS)

for fold, (train_index, test_index) in enumerate(gkf.split(X, y_ra, groups=groups)):
    print(f"\n--- FOLD {fold + 1}/{N_SPLITS} ---")
    X_train, X_test = X[train_index], X[test_index]
    y_train_ra, y_test_ra = y_ra[train_index], y_ra[test_index]

    scaler_X = StandardScaler(); scaler_y_ra = StandardScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)
    y_train_ra_scaled = scaler_y_ra.fit_transform(y_train_ra)

    model_ra = models.Sequential([
        layers.Input(shape=(len(feature_cols),)),
        layers.Dense(32, activation='relu'),
        layers.Dense(64, activation='relu'),
        layers.Dense(32, activation='relu'),
        layers.Dense(1)])
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
    model_ra.compile(loss='mse', optimizer=optimizer)

    print(" -> Training Ra specialist model...")
    model_ra.fit(X_train_scaled, y_train_ra_scaled, epochs=REGRESSOR_EPOCHS, batch_size=BATCH_SIZE,
                 callbacks=[tf.keras.callbacks.EarlyStopping(patience=15, restore_best_weights=True)], verbose=0)

    y_pred_ra_scaled = model_ra.predict(X_test_scaled, verbose=0)
    y_pred_ra_orig = scaler_y_ra.inverse_transform(y_pred_ra_scaled)

    all_true_ra.extend(y_test_ra); all_pred_ra.extend(y_pred_ra_orig)
    # Store the true Rz values that correspond to this fold's test set
    all_true_rz.extend(y_rz[test_index])
    print(" -> Fold complete.")

# ==============================================================================
# Stage 3: Train Rz Assistant Model and Final Evaluation
# ==============================================================================
print("\n\n--- Stage 3: Training Rz Assistant and Final Evaluation ---")
y_true_ra_final = np.array(all_true_ra)
y_pred_ra_final = np.array(all_pred_ra)
y_true_rz_final = np.array(all_true_rz)

# --- Train the simple linear model on the FULL dataset to learn the Ra -> Rz relationship ---
print(" -> Training linear model to learn Ra-Rz correlation...")
linear_assistant = LinearRegression()
linear_assistant.fit(df_final[['Ra']], df_final[['Rz']])

# --- Use the assistant to predict Rz based on the Specialist's Ra predictions ---
y_pred_rz_final = linear_assistant.predict(y_pred_ra_final)

# --- Calculate Overall Performance Metrics ---
r2_ra = r2_score(y_true_ra_final, y_pred_ra_final)
mae_ra = mean_absolute_error(y_true_ra_final, y_pred_ra_final)
r2_rz = r2_score(y_true_rz_final, y_pred_rz_final)
mae_rz = mean_absolute_error(y_true_rz_final, y_pred_rz_final)

print("\n--- Overall Regression Performance (Chained Model) ---")
print(f"Ra Prediction (Specialist) -> R-squared (R): {r2_ra:.4f}, MAE: {mae_ra:.4f}")
print(f"Rz Prediction (Assistant)  -> R-squared (R): {r2_rz:.4f}, MAE: {mae_rz:.4f}")

# --- Visualization ---
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(y_true_ra_final, y_pred_ra_final, alpha=0.7)
plt.plot([y_true_ra_final.min(), y_true_ra_final.max()], [y_true_ra_final.min(), y_true_ra_final.max()], 'r--', lw=2)
plt.title(f'Ra Prediction (Specialist Model) | R = {r2_ra:.3f}')
plt.xlabel("True Ra"); plt.ylabel("Predicted Ra")
plt.grid(True); plt.axis('equal')

plt.subplot(1, 2, 2)
plt.scatter(y_true_rz_final, y_pred_rz_final, alpha=0.7)
plt.plot([y_true_rz_final.min(), y_true_rz_final.max()], [y_true_rz_final.min(), y_true_rz_final.max()], 'r--', lw=2)
plt.title(f'Rz Prediction (from Predicted Ra) | R = {r2_rz:.3f}')
plt.xlabel("True Rz"); plt.ylabel("Predicted Rz")
plt.grid(True); plt.axis('equal')

plt.tight_layout()
plt.show()

print("\nScript finished successfully.")
```
